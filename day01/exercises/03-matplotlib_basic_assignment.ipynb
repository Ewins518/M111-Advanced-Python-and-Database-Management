## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/company_sales_data.csv")
data.shape
```

```python
month_number = data.month_number
total_profit = data.total_profit
data.head()
```

```python
plt.plot(month_number, total_profit)
plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.xticks(month_number)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.

**Expected output**

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
faceCremSales   = data ['facecream']
faceWashSales   = data ['facewash']
toothPasteSales = data ['toothpaste']
bathingsoapSales   = data ['bathingsoap']
shampooSales   = data ['shampoo']
moisturizerSales = data ['moisturizer']

```

```python
data.head()
```

```python
plt.figure(figsize=(10,6))

plt.plot(month_number, faceCremSales,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(month_number, faceWashSales,   label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(month_number, toothPasteSales, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month_number, bathingsoapSales, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month_number, shampooSales, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month_number, moisturizerSales, label = 'ToothPaste Sales Data', marker='o', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(month_number)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

## Exercise 4
**Read toothpaste sales data of each month and show it using a scatter plot**
Also, add a grid in the plot. gridline style should “–“.

**Expected output**

<img src="../data/Tooth_paste_Sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
plt.figure(figsize=(10,5))
plt.scatter(month_number, toothPasteSales, label = 'Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.title(' Tooth paste Sales data')
plt.xticks(month_number)
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
plt.figure(figsize=(10,5))
plt.bar([a-0.25 for a in month_number], faceCremSales, width= 0.25, label = 'Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in month_number], faceWashSales, width= -0.25, label = 'Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(month_number)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
plt.figure(figsize=(10,5))
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [data ['facecream'].sum(), data ['facewash'].sum(), data ['toothpaste'].sum(), 
         data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
plt.figure(figsize=(10,6))
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(month_number, faceCremSales, faceWashSales, toothPasteSales, 
              bathingsoapSales, shampooSales, moisturizerSales, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
